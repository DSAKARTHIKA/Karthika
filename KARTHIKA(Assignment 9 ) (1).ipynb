```python
#2.Clustering methods:
#kmeans clustering 
from sklearn.cluster import KMeans
```

```python
wcss=[]                               
for i in range(1,13):                  
    kmeans=KMeans(n_clusters=i,init= 'k-means++',random_state=42)      
    kmeans.fit(data)           
    wcss.append(kmeans.inertia_)   
```

```python
plt.plot(range(1,13),wcss)
plt.title('The Elbow Method')
plt.xlabel('No of Clusters')
plt.ylabel('WCSS')
plt.show()                 # found 3 as optimum no of clusters,as it is at elbow
```

```python
kmeans=KMeans(n_clusters=3,init= 'k-means++',random_state=42)
y_kmeans=kmeans.fit_predict(data)
```

```python
y_kmeans 
```

```python
from sklearn.metrics import silhouette_score
```

```python
sil_kmeans=silhouette_score(data,y_kmeans)
```

```python
sil_kmeans
```

```python
#Agglomerative clusterting
import scipy.cluster.hierarchy as sch
```

```python
dendrogram=sch.dendrogram(sch.linkage(data,method='ward'))

plt.title('Dendrogram')
plt.xlabel('Data Points')
plt.ylabel('Euclidean Distance')
plt.show()                    #optimum no of clusters is 3
```

```python
from sklearn.cluster import AgglomerativeClustering
```

```python
ahc=AgglomerativeClustering(n_clusters=3,affinity='euclidean',linkage='ward')
y_ahc=ahc.fit_predict(data)
```

```python
y_ahc
```

```python
sil_ahc=silhouette_score(data,y_ahc)
```
